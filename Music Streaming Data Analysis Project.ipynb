Churn = 0, meaning the average skip_rate for remaining users is ≈ 0.15

Churn = 1, meaning the average skip_rate for users who left is ≈ 0.55

In this case, we can say that users who left the platform skipped songs much more frequently, while remaining users listened to the content more often.

In other words, a high skip_rate is a key indicator that increases the risk of churn.

```python
df.groupby("device").agg(
user_count=("user_id","count"),
avg_daily_minutes=("daily_minutes","mean"),
avg_skip_rate=("skip_rate","mean")
)
```

In contrast, desktop users spend more time on the platform and have lower skip rates.

```python
df.groupby("subscription_type").agg(
user_count=("user_id","count"),
total_daily_minutes=("daily_minutes","sum"),
avg_daily_minutes=("daily_minutes","mean"),
avg_skip_rate=("skip_rate","mean")
)
```

When we analyze the data by subscription type, the results show that premium users have higher total and average listening times and a lower song-skipping rate.

On average, premium users consume three times as much content.

Free users, on the other hand, exhibit lower engagement and a higher tendency to churn.


```python
df.groupby("country").agg(
user_count=("user_id","count"),
avg_daily_minutes=("daily_minutes","mean"),
avg_skip_rate=("skip_rate","mean")
).sort_values(by="avg_daily_minutes",ascending=False)
```

```python
df["subscription_type"].value_counts().plot(kind="bar", color=["#4CAF50", "#FF9800"])
plt.title("Subscription Type by User Count")
plt.xlabel("Subscription Type")
plt.ylabel("User Count")
plt.show()
```

When we look at the graph showing the distribution of subscriptions in the dataset, we see that there are more Free users than Premium users.

```python
df.groupby("device")["daily_minutes"].mean().plot(
    kind="bar",
    color=["#3F51B5", "#009688"]
)

plt.title("Average Daily Minutes by Device")
plt.xlabel("Device")
plt.ylabel("Average Daily Minutes")
plt.show()
```

An analysis of the graph shows that desktop users listen to music for longer periods than mobile users.

In this context, we can conclude that desktop users engage with the platform in a more “focused” manner and during longer sessions, while mobile users exhibit shorter and more fragmented usage patterns.

```python
df.groupby("churned")["skip_rate"].mean().plot(
    kind="bar",
    color=["#2ECC71", "#E74C3C"]  # 0 = kalma (yeşil), 1 = churn (kırmızı)
)

plt.title("Average Skip Rate by Churn Status")
plt.xlabel("Churned (0 = Stayed, 1 = Churned)")
plt.ylabel("Average Skip Rate")
plt.show()
```

Upon examining the graph, it is evident that the average skip_rate for users who have left the platform (churned = 1) is significantly higher than that of remaining users (churned = 0).

```python
df["daily_minutes"].plot(kind="hist",bins=10)
plt.title("Daily Minutes Distribution")
plt.xlabel("Daily Minutes")
plt.show()
```

Looking at the distribution of daily listening time, we can see that the majority of users engage in low to moderate listening, while a small number of users show high usage.

The right-skewed distribution indicates that a few high-value users are skewing the average.

```python
df.plot(kind="scatter",x="skip_rate",y="daily_minutes")
plt.title("Skip Rate vs Daily Minutes")
plt.xlabel("Skip Rate")
plt.ylabel("Daily Minutes")
plt.show()
```

When we examine the relationship between the skip rate and daily listening time using a scatter plot, we see that as the skip rate increases, users’ daily listening time tends to decrease.

This indicates a negative correlation between user satisfaction and the time spent on the platform.

```python
df.groupby("churned").agg(
user_count=("user_id","count"),
avg_skip_rate=("skip_rate","mean"),
avg_daily_minutes=("daily_minutes","mean")
)
```

When we analyze user behavior by grouping it based on churn status, the results show that users who leave the platform have a higher song-skipping rate and a lower daily listening time.

```python
df[df["churned"]==1].groupby("country").agg(
churned_users=("user_id","count")
).sort_values(by="churned_users",ascending=False)
```

When we look at the results by country for users who have left the platform, the data shows that Turkey has the highest churn rate, while Germany has the lowest.

In this context, we can say that user churn varies by geographic region.

At the same time, we can conclude that strategies must be optimized on a country-by-country basis to minimize the impact of this trend.

Based on the results, we can say that the highest average skip rate is generally found in the Hip Hop and Pop genres.

In the Classical and Jazz genres, however, the skip rate is lower.
